```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('/content/drive/My Drive/YIP/data_ayurveda/lotus/l1.jpg')
min_int = np.min(img)
max_int = np.max(img)
streched =((img-min_int)/(max_int-min_int)*255).astype(np.uint8)
```

```python
hist_o = cv2.calcHist([img],[0],None,[255],[0,255])
hist_s = cv2.calcHist([streched],[0],None,[255],[0,255])
```

```python
img1 = cv2.imread('/content/drive/My Drive/YIP/data_ayurveda/lotus/l1.jpg',0)
em = cv2.equalizeHist(img1)
hist_e = cv2.calcHist([em],[0],None,[255],[0,255])
```

```python
plt.figure(figsize=(12,8))
```

```python
plt.subplot(231)
plt.title('Original Image')
plt.imshow(img,'Accent')
plt.axis('off')
plt.subplot(232)
plt.title('Streched Image')
plt.imshow(streched,'Accent')
plt.axis('off')
plt.subplot(233)
plt.title('Equilized Image')
plt.imshow(em,'Accent')
plt.axis('off')
plt.subplot(234)
plt.title('Equilized Image')
plt.imshow(hist_o)
plt.xlim([0,255])
plt.subplot(235)
plt.title('Equilized Image')
plt.imshow(hist_s)
plt.xlim([0,255])
plt.subplot(236)
plt.title('Equilized Image')
plt.imshow(hist_e)
plt.xlim([0,255])
```
